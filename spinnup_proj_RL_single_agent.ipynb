OpenAI Spinning Up: https://spinningup.openai.com/en/latest/user/running.html

```python
import os
from spinup import ppo_pytorch as ppo
import torch
import gym
import rsoccer_gym

env_fn = lambda : gym.make('SSLPassEndurance-v0')

ac_kwargs = dict(hidden_sizes=[64,64], activation=torch.nn.ReLU)

parent_dir = os.getcwd()
filepath = 'path/to/output_dir'
path = os.path.join(parent_dir, filepath)
if not os.path.isdir(path):
  print("here")
  os.makedirs(path)

logger_kwargs = dict(output_dir=filepath, exp_name='experiment_0')

ppo(env_fn=env_fn, ac_kwargs=ac_kwargs, steps_per_epoch=1500, epochs = 1000, logger_kwargs=logger_kwargs)
```

Carreggar modelos e resultados: https://spinningup.openai.com/en/latest/user/saving_and_loading.html

```python
import io
import pandas as pd
import matplotlib.pyplot as plt

with open(filepath + "/progress.txt") as f:
    fileobject = io.StringIO(f.read().replace('""','"'))

csv = pd.read_csv(fileobject, sep='\t',  lineterminator='\n', names=None)
print(csv)
plt.figure()

x = csv['Epoch']
y = csv['AverageEpRet']

plt.plot(x,y)
```

```python
# restart the kernel: ambiente de execução > reiniciar ambiente de execução

import gym
import rsoccer_gym

from PIL import Image
import matplotlib.pyplot as plt
from IPython import display as ipythondisplay

from pyvirtualdisplay import Display
display = Display(visible=0, size=(200, 150))
display.start()

# Using VSS Single Agent env
env = gym.make('SSLPassEndurance-v0')

# Run for 1 episode and print reward at the end
for i in range(5):
    env.reset()
    frames = []
    done = False
    while not done:
        # Step using random actions
        action = env.action_space.sample()
        next_state, reward, done, _ = env.step(action)
        screen = Image.fromarray(env.render(mode='rgb_array'))
        screen = screen.resize(size=(700, 550))
        frames.append(screen)
    print("ep." + str(i+1) + " " + str(reward))

env.close()
```
